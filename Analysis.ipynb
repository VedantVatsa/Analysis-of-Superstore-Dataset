```python
# Load necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df=pd.read_csv("SampleSuperstore.csv")
df.head()
```

```python
# To remove a column
df.drop(columns="Postal Code", inplace=True)

df.head()
```

```python
# To find unique values in the given column
print(df["Ship Mode"].unique())
print(df["Segment"].unique())
print(df["Country"].unique())
print(df["Category"].unique())
print(df["Sub-Category"].unique())
print(df["Region"].unique())
```

```python
# Statistical description of the data
df.describe()
```

```python
# To show non-null count and data type
df.info()
```

```python
#To find any missing value
df.isna().sum()
```

```python
# Sales analysis based on Region

df.groupby("Region")["Sales"].sum().plot.bar()
```

```python
# Profit analysis based on Region

df.groupby("Region")["Profit"].sum().plot.bar()
```

```python
# Sales analysis based on Region (Pie chart)

df.groupby("Region")["Sales"].sum().plot.pie(autopct="%1.0f%%")
```

```python
# Profit analysis based on Region

df.groupby("Region")["Profit"].sum().plot.pie(autopct="%1.0f%%")
```

```python
# Sales analysis based on Segment

df.groupby("Segment")["Sales"].sum().plot.bar()
```

```python
# Profit analysis based on Segment

df.groupby("Segment")["Profit"].sum().plot.bar()
```

```python
# Sales analysis based on Category

df.groupby("Category")["Sales"].sum().plot.bar()
```

```python
# Profit analysis based on Category

df.groupby("Category")["Profit"].sum().plot.bar()
```

```python
# Sales analysis based on Category Pie chart
 
df.groupby("Category")["Sales"].sum().plot.pie(autopct="%1.0f%%")
```

```python
# Profit analysis based on Category Pie chart
 
df.groupby("Category")["Profit"].sum().plot.pie(autopct="%1.0f%%")
```

```python
# Profit analysis based on State
 
df.groupby("State")["Profit"].sum().plot.bar()
```
